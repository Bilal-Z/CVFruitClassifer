```python
import cv2
import matplotlib.pyplot as plt
```

```python
partial = cv2.imread('../DATA/fruits-360_dataset/fruits-360/T&T/Test/Apple Braeburn/36_100.jpg')
partial = cv2.cvtColor(partial,cv2.COLOR_BGR2RGB)
plt.imshow(partial)
```

```python
full = cv2.imread('../DATA/fruits-360_dataset/fruits-360/test-multiple_fruits/pomegranate.jpg')
full = cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
plt.imshow(full)
```

```python
full_copy = full.copy()
```

```python
res = cv2.matchTemplate(full_copy,partial,method=cv2.TM_CCOEFF_NORMED)
```

```python
height, width,channels = partial.shape
```

```python
height
```

```python
# Grab the Max and Min values, plus their locations
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
# Set up drawing of Rectangle
   
top_left = max_loc
      
# Assign the Bottom Right of the rectangle
bottom_right = (top_left[0] + width, top_left[1] + height)

# Draw the Red Rectangle
cv2.rectangle(full_copy,top_left, bottom_right, 255, 10)

# Plot the Images
plt.subplot(121)
plt.imshow(res)
plt.title('Result of Template Matching')
    
plt.subplot(122)
plt.imshow(full_copy)
plt.title('Detected Point')
plt.suptitle('cv2.TM_CCORR')
    
    
plt.show()
```
